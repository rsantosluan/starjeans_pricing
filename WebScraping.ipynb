## 1. Import

```python
import pandas as pd
from datetime import datetime


#Imports HTML
import requests
from bs4 import BeautifulSoup

```

## 2. Data extraction

```python
url = 'https://www2.hm.com/en_us/men/products/jeans.html'

header = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) \
    AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'
}
page = requests.get(url, headers=header)
```

```python
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
products = soup.find('ul', 'products-listing small')
```

```python
#List of product ID and category
product_list = products.find_all('article', class_='hm-product-item')
#List of product price
product_price = products.find_all('span', class_='price regular')
#List of product name
product_name = products.find_all('a', class_='link')
```

```python
data = pd.DataFrame()
data['prod_id'] = [p.get('data-articlecode') for p in product_list]
data['scrapy_datetime'] = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
data['prod_name'] =[p.get_text() for p in product_name]
data['prod_category'] = [p.get('data-category') for p in product_list]
data['prod_price'] = [float(p.get_text()[2:]) for p in product_price]


```
